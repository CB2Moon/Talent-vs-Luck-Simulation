```python
import numpy as np
from scipy.stats import truncnorm
import matplotlib.pyplot as plt
%matplotlib inline

LUCKY=1
UNLUCKY=0

class Person:
    def __init__(self, position, talent, wealth=10):
        self.position = np.array(position)
        self.talent = talent
        self.wealth = wealth
        self.wealth_history = [wealth]
        self.luck_count = 0
        self.unluck_count = 0

class Event:
    def __init__(self, event_type, position, time):
        self.event_type = event_type
        self.position = np.array(position)
        self.time = time

class Environment:
    def __init__(self, dimension, size, num_people, talent_distribution, wealth_increment_factor, wealth_decrement_factor, salary, num_years, steps_per_year, num_events, event_ratio, event_distribution_periodic, event_period, event_interval):
        self.dimension = dimension                      # 2 or 3
        self.size = size                                # square, cube
        self.num_people = num_people
        self.talent_distribution = talent_distribution  # (mean, std_dev)
        self.wealth_increment_factor = wealth_increment_factor
        self.wealth_decrement_factor = wealth_decrement_factor
        self.salary = salary
        self.num_years = num_years
        self.steps_per_year = steps_per_year
        self.num_events = num_events
        self.event_ratio = event_ratio                  # default lucky event ratio
        self.event_distribution_periodic = event_distribution_periodic  # True for periodic, false for random
        self.event_period = event_period                # in year
        self.event_interval = event_interval            # interval used if periodic
        self.validate_parameters()
        self.people = self.init_people()
        self.events = self.init_events()

    def validate_parameters(self):
        assert 0 <= self.wealth_decrement_factor <= 1, "`wealth_decrement_factor` must be a number between 0 and 1"
        assert isinstance(self.event_ratio, (int, float)), "`event_ratio` must be a number between 0 and 1"
        assert 0 <= self.event_ratio <= 1, "`event_ratio` must be a number between 0 and 1"

        assert isinstance(self.event_period, (int, float)), "`event_period` must be a positive number"
        assert self.event_period > 0, "`event_period` must be a positive number"

        assert isinstance(self.event_interval, tuple), "`event_interval` must be a tuple of two numbers"
        assert len(self.event_interval) == 2, "`event_interval` must be a tuple of two numbers"
        assert all(isinstance(n, (int, float)) for n in self.event_interval), "`event_interval` must be a tuple of two numbers"
        assert self.event_interval[0] < self.event_interval[1], "`event_interval` must be a tuple of two numbers, the first less than the second"
    
    def init_people(self):
        people = []
        lower, upper = 0, 1  # lower and upper bounds for talent
        mu, sigma = self.talent_distribution  # mean and standard deviation
        distribution = truncnorm((lower - mu) / sigma, (upper - mu) / sigma, loc=mu, scale=sigma)
        
        for _ in range(self.num_people):
            position = [np.random.uniform(0, self.size) for _ in range(self.dimension)]
            talent = distribution.rvs(1)[0]
            people.append(Person(position, talent))
        return people
    
    def init_events(self):
        events = []
        for _ in range(self.num_events):
            event_type = np.random.choice([LUCKY, UNLUCKY], p=[self.event_ratio, 1-self.event_ratio])
            position = [np.random.uniform(0, self.size) for _ in range(self.dimension)]
            time = np.random.randint(0, self.num_years * self.steps_per_year)
            events.append(Event(event_type, position, time))
        # Sort the events by the time they occur.
        events.sort(key=lambda e: e.time)
        return events
    
    def init_events(self):
        events = []
        self.lucky_counts = [0] * (self.num_years * self.steps_per_year)
        self.unlucky_counts = [0] * (self.num_years * self.steps_per_year)
        
        for _ in range(self.num_events):
            position = [np.random.uniform(0, self.size) for _ in range(self.dimension)]
            time = np.random.randint(0, self.num_years * self.steps_per_year)
            
            # Calculate the oscillating event ratio based on the current time.
            if self.event_distribution_periodic:
                t = time // self.steps_per_year
                ratio = (self.event_interval[1] - self.event_interval[0]) / 2 * np.sin(2 * np.pi / self.event_period * t) + (self.event_interval[0] + self.event_interval[1]) / 2
            else:  # random
                ratio = self.event_ratio
                
            event_type = np.random.choice([LUCKY, UNLUCKY], p=[ratio, 1-ratio])

            events.append(Event(event_type, position, time))
            if event_type == LUCKY:
                self.lucky_counts[time] += 1
            else:
                self.unlucky_counts[time] += 1

        # Sort the events by the time they occur.
        events.sort(key=lambda e: e.time)
        return events

    def distance(self, pos1, pos2):
        return np.linalg.norm(pos1 - pos2)
    
    def run(self):
        event_index = 0
        for t in range(self.num_years * self.steps_per_year):
            # Only loop over events that occur at the current time.
            while event_index < len(self.events) and self.events[event_index].time == t:
                event = self.events[event_index]
                for person in self.people:
                    if self.distance(person.position, event.position) <= 5:
                        if event.event_type == LUCKY:
                            if np.random.uniform() <= person.talent:  # if grasp the chance
                                person.luck_count += 1
                                person.wealth = (1+self.wealth_increment_factor) * person.wealth
                        else:  # unlucky
                            person.unluck_count += 1
                            person.wealth = (1-self.wealth_decrement_factor) * person.wealth
                    person.wealth += self.salary
                event_index += 1
            
            for person in self.people:
                person.wealth_history.append(person.wealth)
        self.people.sort(key=lambda p: p.wealth)
    
    def plot_environment(self):
        # Plot people
        x = [person.position[0] for person in self.people]
        y = [person.position[1] for person in self.people]
        plt.scatter(x, y, marker='2', color='black', alpha=0.9, label='People')

        # Plot events
        for event in self.events:
            if event.event_type == LUCKY:
                plt.scatter(event.position[0], event.position[1], color='blue', marker='P', alpha=0.3, label='Lucky Event')
            else:
                plt.scatter(event.position[0], event.position[1], color='red', marker='x', alpha=0.7, label='Unlucky Event')

        # Handle legends
        handles, labels = plt.gca().get_legend_handles_labels()
        by_label = dict(zip(labels, handles))
        plt.legend(by_label.values(), by_label.keys())

        plt.xlabel('X')
        plt.ylabel('Y')
        plt.title('Environment')
        plt.show()
    
    def plot_wealth_distribution(self):
        wealths = [person.wealth for person in self.people]
        plt.hist(wealths, bins=100, edgecolor="white")
        plt.xlabel('Wealth')
        plt.ylabel('Number of People')
        plt.show()
        
    def plot_talent_distribution(self):
        talents = [person.talent for person in self.people]
        plt.hist(talents, bins=50, range=(0, 1), edgecolor="white")
        plt.xlabel('Talent')
        plt.ylabel('Number of people')
        plt.show()

    def plot_talent_vs_wealth(self):
        talents = [person.talent for person in self.people]
        wealths = [person.wealth for person in self.people]
        plt.stem(talents, wealths)
        plt.xlabel('Talent')
        plt.ylabel('Wealth')
        plt.show()

    def plot_wealth_over_time(self, portion, segment=None):
        assert 0 < portion < 1, "`portion` must be between 0 and 1"
        if segment == 'richest':
            people_to_plot = self.people[int((1-portion) * len(self.people)):]
        elif segment == 'poorest':
            people_to_plot = self.people[:int(portion * len(self.people))]
        elif segment == 'middle':
            people_to_plot = self.people[int((0.5-portion/2) * len(self.people)):int((0.5+portion/2) * len(self.people))]
        else:
            people_to_plot = self.people

        for person in people_to_plot:
            plt.plot(person.wealth_history)
        
        plt.xlabel('Time')
        plt.ylabel('Wealth')
        plt.title(f'Wealth over time ({segment if segment else "all"} people)')
        plt.show()
        
    def plot_event_counts(self):
        plt.plot(self.lucky_counts, label='Lucky')
        plt.plot(self.unlucky_counts, label='Unlucky')
        plt.xlabel('Time')
        plt.ylabel('Number of Events')
        plt.legend()
        plt.show()
        
    def plot_wealth_vs_talent(self, scale='linear', base_range=10):
        assert scale in ['linear', 'logarithmic'], "scale must be either 'linear' or 'logarithmic'"
        
        if scale == 'linear':
            # Calculate the bin edges for the linear scale
            min_wealth = min(person.wealth for person in self.people)
            max_wealth = max(person.wealth for person in self.people)
            bin_edges = np.arange(min_wealth, max_wealth + base_range, base_range)
        else:  # logarithmic
            # Calculate the bin edges for the logarithmic scale
            min_wealth = np.log2(min(person.wealth for person in self.people))
            max_wealth = np.log2(max(person.wealth for person in self.people))
            bin_edges = np.logspace(min_wealth, max_wealth, base=base_range)

        # Place each person in the appropriate bin and plot their talent
        for person in self.people:
            bin_index = np.digitize(person.wealth, bin_edges) - 1
            plt.stem([bin_edges[bin_index]], [person.talent])

        plt.xlabel('Wealth')
        plt.ylabel('Talent')
        plt.title('Wealth vs Talent')
        plt.show()
```

```python
%matplotlib inline
env = Environment(
    dimension=2,
    size=201,
    num_people=1000,
    talent_distribution=(0.6, 0.1),
    wealth_increment_factor=1,
    wealth_decrement_factor=0.5,
    salary=0,
    num_years=40,
    steps_per_year=2,
    num_events=500,
    event_ratio=0.5,
    event_distribution_periodic=False,
    event_period=12,
    event_interval=(0.2, 0.8)
)
env.run()
```
